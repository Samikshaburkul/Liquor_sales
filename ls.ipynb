```python
import pandas as pd
df = pd.read_csv('/Users/samikshaburkul/Desktop/Datawarehousing project/Liquor_Sales.csv')
print(df.dtypes)
df.head()
```

```python
df.dropna(inplace=True)
df.columns = ['invoice_number','date','store_number','store_name','address','city','zip_code','store_location','county_number','county','category','category_name','vendor_number','vendor_name', 'item_number','item_description','pack','bottle_volume_ml','state_bottle_cost','state_bottle_retail','bottles_sold','sale_usd','volume_sold_liters','volume_sold_gallons']
df.drop('item_number', axis=1, inplace=True)
df['date'] = pd.to_datetime(df['date'], format='%m/%d/%Y').dt.strftime('%Y-%m-%d')
df['store_number'] = df['store_number'].astype(int)
df['county_number'] = df['county_number'].astype(int)
df['category'] = df['category'].astype(int)
df['vendor_number'] = df['vendor_number'].astype(int)
df['pack'] = df['pack'].astype(int)
df['bottle_volume_ml'] = df['bottle_volume_ml'].astype(int)
df['bottles_sold'] = df['bottles_sold'].astype(int)
df.head()
```

```python
df.to_csv('liquor_sales_updated.csv', index=False)
```
